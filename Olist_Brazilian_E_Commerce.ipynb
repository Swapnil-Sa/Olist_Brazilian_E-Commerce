```python
def inspect_table(df, name):
    print(f"\n{'='*50}")
    print(f"📊 DATASET: {name}")
    print(f"{'='*50}")

    print(f"\n🔹 Shape: {df.shape}")

    print("\n🔹 Data Types:")
    print(df.dtypes)

    print("\n🔹 Missing Values:")
    print(df.isnull().sum().sort_values(ascending=False))

    duplicate_count = df.duplicated().sum()
    print(f"\n🔹 Duplicate Rows: {duplicate_count}")

    print("\n🔹 Unique Values:")
    print(df.nunique())

    print("\n🔹 Sample Data:")
    print(df.head(3))
```

```python
orders = pd.read_csv("/content/olist_orders_dataset.csv")
customers = pd.read_csv("/content/olist_customers_dataset.csv")
order_items = pd.read_csv("/content/olist_order_items_dataset.csv")
payments = pd.read_csv("/content/olist_order_payments_dataset.csv")
reviews = pd.read_csv("/content/olist_order_reviews_dataset.csv")
products = pd.read_csv("/content/olist_products_dataset.csv")
sellers = pd.read_csv("/content/olist_sellers_dataset.csv")
category = pd.read_csv("/content/product_category_name_translation.csv")
geolocation = pd.read_csv("/content/olist_geolocation_dataset.csv")
```

```python
inspect_table(orders, "Orders")
inspect_table(customers, "Customers")
inspect_table(order_items, "Order Items")
inspect_table(payments, "Payments")
inspect_table(reviews, "Reviews")
inspect_table(products, "Products")
inspect_table(sellers, "Sellers")
inspect_table(category, "Category Translation")
inspect_table(geolocation, "Geolocation")
```

```python
for name, df in {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "payments": payments,
    "reviews": reviews,
    "products": products,
    "sellers": sellers
}.items():
    print(f"\n{name.upper()}")
    print(df.columns.tolist())
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```

```python
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
).dt.days
```

```python
orders['is_late'] = (
    orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date']
).astype(int)
```

```python
orders['processing_days'] = (
    orders['order_approved_at'] - orders['order_purchase_timestamp']
).dt.days
```

```python
orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
orders['order_day'] = orders['order_purchase_timestamp'].dt.day
```

```python
payments_agg = payments.groupby('order_id').agg({
    'payment_value': 'sum',
    'payment_installments': 'max'
}).reset_index()
```

```python
reviews_clean = reviews[['order_id', 'review_score']]
```

```python
products['product_category_name'] = products['product_category_name'].fillna('Unknown')
```

```python
orders.to_csv("clean_orders.csv", index=False)
payments_agg.to_csv("clean_payments.csv", index=False)
reviews_clean.to_csv("clean_reviews.csv", index=False)
products.to_csv("clean_products.csv", index=False)
```
